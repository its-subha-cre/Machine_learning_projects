```python
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
iris
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
df['target']=iris.target
```

```python
df
```

```python
iris.target_names
```

```python
df['flower_name'] = df['target'].apply(lambda x: 'setosa' if x == 0 else 'versicolor' if x == 1 else 'virginica')
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
```

```python
s=df
```

```python
s.info()
```

```python
y=s.target
```

```python
X=s.drop(['petal length (cm)','petal width (cm)','target','flower_name'],axis='columns')
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import cross_val_Score
c=cross_val_score
```

```python
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
c=cross_val_score(KNeighborsClassifier(),X,y,cv=5)
c.mean()
```

```python
knn.predict(X_test)
```

```python
knn.score(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
bag_model=BaggingClassifier(
    estimator=KNeighborsClassifier(),
    n_estimators=10,
    max_samples=0.8)
bag_model.fit(X_train,y_train)
```

```python
cross_val_score(BaggingClassifier(),X,y,cv=5).mean()
```

```python
bag_model.score(X_test,y_test)
```
