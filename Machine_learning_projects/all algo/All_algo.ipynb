```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
dir(wine)
```

```python
df=pd.DataFrame(wine.data,columns=wine.feature_names)
```

```python
wine.feature_names
```

```python
df
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(0.95)
```

```python
df_pca=pca.fit_transform(df)
```

```python
df
```

```python
df['target']=wine.target
```

```python
wine.target_names
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
a=cross_val_score(RandomForestClassifier(n_estimators=100),df_pca,df.target,cv=5).mean()
b=cross_val_score(LogisticRegression(),df_pca,df.target,cv=5).mean()
c=cross_val_score(DecisionTreeClassifier(),df_pca,df.target,cv=5).mean()
d=cross_val_score(SVC(),df_pca,df.target,cv=5).mean()
e=cross_val_score(KNeighborsClassifier(n_neighbors=15),df_pca,df.target,cv=5).mean()
f=cross_val_score(GaussianNB(),df_pca,df.target,cv=5).mean()
l=[a,b,c,d,e,f]
```
