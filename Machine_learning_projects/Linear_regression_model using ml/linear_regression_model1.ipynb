```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
df.head(7)
```

```python
df['per capita income (US$)']=df['per capita income (US$)'].astype('int64')
```

```python
df.info()
```

```python
df.rename(columns={'per capita income (US$)':'capital'},inplace=True)
df
```

```python
x=df['year'].values.reshape(1,-1)
y=df.capital.values.reshape(1,-1)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('capita_income')
plt.title('Scatter plot')
plt.scatter(df['year'],df['capital'])
plt.savefig('scatter_plot.png')
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df['capital'])
```

```python
df.shape
```

```python
new_year=np.array([[2016]])
reg.predict(new_year)
```

```python
m=plt.plot(df['year'],df['capital'])
plt.savefig('linefig.png')
```

```python
l=[]
l.append(reg.coef_)
```

```python
l.append(reg.intercept_)
```
